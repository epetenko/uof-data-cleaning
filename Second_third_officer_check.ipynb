```python
sec_of['Is_second'] = 'True'
```

```python
th_of = th_of.rename(columns={u'Officer_3_first':'Officer_1_first', 
        u'Officer_3_last':'Officer_1_last', u'office3id':'officer1id', u'Officer_3_rank':'Officer_1_rank', 
        u'Officer_3_exp':'Officer_1_exp', u'Officer_3_sex':'Officer_1_sex',
       u'Officer_3_race':'Officer_1_race', u'Officer_3_badge_no':'Officer_1_badge_no', 
        u'Officer_3_injured':'Officer_1_injured',
       u'Officer_3_hosp':'Officer_1_hosp', u'Officer_3_killed':'Officer_1_killed'})
```
