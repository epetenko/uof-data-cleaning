```python
df.Subject_2_charges = df.Subject_2_charges.str.upper()
charges2 = pd.DataFrame(df.Subject_2_charges.str.split(', ', expand=True))
charges2[0] = charges2[0].map(standard_charge)
charges2[1] = charges2[1].map(standard_charge)
charges2[2] = charges2[2].map(standard_charge)
charges2[3] = charges2[3].map(standard_charge)
charges2[4] = charges2[4].map(standard_charge)
charges2[5] = charges2[5].map(standard_charge)
charges2[6] = charges2[6].map(standard_charge)
charges2 = charges2.fillna('')
charges2['New_charges2'] = charges2[0].astype(str) + ", " + charges2[1].astype(str) + ', ' + charges2[2].astype(str) + ', ' + charges2[3].astype(str)+ ', ' + charges2[4].astype(str)+ ', ' + charges2[5].astype(str) + ', ' +charges2[6].astype(str)
ch_merge2 = pd.concat([df.Subject_2_charges, charges2], axis=1)

ch_merge2 = ch_merge2.set_index('Subject_2_charges')

ch_dict2 = ch_merge2['New_charges2'].to_dict()

df['Subject_2_charges_st'] = df.Subject_2_charges.map(ch_dict2)
```
